```python
import networkx as nx 
import matplotlib.pyplot as plt

# iniciar grafo 
G3 = nx.Graph()

# cria um dicionario com as posições [ opcional , apenas para visualização]
pos = {}
pos[1]= (1000,1500);pos[2]=(1000,500);pos[3]=(500,0);
pos[4]= (0,500);pos[5]=(0,1500);pos[6]=(500,2000);
pos[7]= (1000,2500);pos[8]=(0,2500);pos[9]=(500,3500)

# add arestas pelas referencias dos nos 
G3.add_edge(1,2);G3.add_edge(1,4);G3.add_edge(1,6);
G3.add_edge(2,5);G3.add_edge(2,3);G3.add_edge(2,4);
G3.add_edge(3,4);G3.add_edge(4,5);G3.add_edge(5,6);
G3.add_edge(6,7);G3.add_edge(6,8);G3.add_edge(7,8);
G3.add_edge(7,9);G3.add_edge(8,9)

# desenha o grafo 
plt.figure(figsize=(14,9))
nx.draw_networkx(G3,pos)
 
```
